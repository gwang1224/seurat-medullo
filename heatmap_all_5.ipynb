```python
import csv
import pandas as pd
```

```python
files = [
    'subcluster0',
    'subcluster1',
    'subcluster2',
    'subcluster3',
    'subcluster4',
    'subcluster5',
    'subcluster6',
    'subcluster7',
    'subcluster8'
]

for file in files:
    
    gene_list = []
    cell_type_list = []
    data_dict = {}

    file_name = file + ".csv"
    path = '/Users/gracewang/Documents/mesirov_lab/seurat-medullo/panglao_subclusters' + file_name
    with open(path, 'r') as tsv_file:
        tsv_reader = csv.reader(tsv_file, delimiter='\t')
        data = list(tsv_reader)

    for sample in data:
        # print(sample)
        if len(sample) == 7:
            gene_list.append(sample[1])
            cell_type_list.append(sample[5])
    
    gene_list = list(set(gene_list))
    cell_type_list = list(set(cell_type_list))


    #creates a dictionary with key = gene and value = 0
    gene_dict = {gene: 0 for gene in gene_list}
    # print(gene_dict)

    

    # creates dictionary of dictionaries with key as cell type and gene count as value
    for cell in cell_type_list:
        data_dict[cell] = gene_dict.copy()
    
    # print(data_dict)


    for sample in data:
        # print(sample)
        if len(sample) == 7:
            if int(sample[6]) <= 5:
                data_dict[sample[5]][sample[1]] += 1
    
    samples_to_delete = list(data_dict.keys())

    for sample in samples_to_delete:
        all_zero = all(value == 0 for value in data_dict[sample].values())
        if all_zero:
            del data_dict[sample]

    # print(data_dict)

    df = pd.DataFrame(data_dict)
    df = df.transpose()

    df_file_name = "df_" + file + ".csv"

    df.to_csv(df_file_name, index=True)
```

```python
import csv

files = [
    'df_subcluster0.csv',
    'df_subcluster1.csv',
    'df_subcluster2.csv',
    'df_subcluster3.csv',
    'df_subcluster4.csv',
    'df_subcluster5.csv',
    'df_subcluster6.csv',
    'df_subcluster7.csv',
    'df_subcluster8.csv'
]

for file in files:
    path = '/Users/gracewang/Documents/mesirov_lab/seurat-medullo/df_data_5/' + file
    # Create a list to store rows you want to keep
    rows_to_keep = []

    # Open the file in read mode
    with open(path, 'r') as file:
        # Create a CSV reader object
        csv_reader = csv.reader(file)
        
        # Iterate over each row in the CSV file
        for row in csv_reader:
            if row[0] != "Unknown":
                # Only add rows where index 0 is not "Unknown"
                rows_to_keep.append(row)

    # Write the remaining rows back to the CSV file
    with open(path, 'w', newline='') as file:
        # Create a CSV writer object
        csv_writer = csv.writer(file)
        
        # Write each row to the file
        for row in rows_to_keep:
            csv_writer.writerow(row)

```
